# データ概要確認
 SIGNATEの【練習問題】銀行の顧客ターゲティングのデータ( https://signate.jp/competitions/1/data )を使用して, train.csvの概要を確認する.

## Import

```python
import numpy as np
import pandas as pd

# 可視化ライブラリ
import seaborn as sns
import matplotlib.pyplot as plt
```

## Setting

```python
# Path
input_path = "../input_data/"

# Set Display Max Columns
pd.set_option("display.max_columns", 50)

# Figure Size
plt.rcParams["figure.figsize"] = (10, 6)
```

## Read Data
 csvファイルを読込む.

```python
train = pd.read_csv(
    input_path + "bank/train.csv", # データ
    sep=",",                       # 区切り文字の指定
    header=0,                      # ヘッダーは何行目か
    quotechar="\""                 # 囲み文字の指定
)
```

## Overview
 データの最初の20行を確認.

```python
train.head(20)
```

 各変数の型確認.

```python
train.dtypes
```

 変数の各種統計量確認.

```python
train.describe()
```

## 単変数の分布確認(連続変数)
 ヒストグラムで分布確認をする.

```python
sns.histplot(
    train["age"],    # データ
    kde = True,      # 近似密度関数の表示有無
    bins = 20        # 変数の刻み数
)
```

## 単変数の分布確認(質的変数)

```python
sns.countplot(data=train, x="job")
plt.xticks(rotation=30)
```

## 目的変数とのクロス集計(連続変数)

データ分析では、目的変数と特徴量の関係を見ることが大事.  
今回は目的変数がカテゴリなので、カテゴリ変数ごとに特徴量を可視化する.

```python
# 離散化
tmp = pd.cut(train["balance"], bins=10) # binsの整数値で最大値と最小値の間を等間隔で分割

# クロス集計
dat_cross = pd.crosstab(tmp, train["y"])
dat_cross
```

## 目的変数とのクロス集計(カテゴリ変数)

```python
pd.crosstab(train["marital"], train["y"])
```

heatmapを使用するとよりわかりやすくなる.

```python
train["age10"] = train["age"]//10 * 10
sns.heatmap(
    pd.crosstab(train["job"], train["age10"]),
    annot=True,
    fmt='d'
)
```

## 目的変数ごとのヒストグラム/箱ひげ図

カテゴリ変数ごとの連続変数の関係は、ヒストグラムや箱ひげ図を使用すると分布の違い等が確認できる.

```python
# 箱ひげ図
sns.boxplot(data=train, x="y", y="age")
```

```python
# ヒストグラム
sns.histplot(data=train, x="age",hue="y",bins=20)
```
