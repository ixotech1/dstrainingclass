## 1. 住宅ローンと個人ローンの契約有無のパターンごとに年間平均残高の平均と定額預金申込率を知りたい

```python
table_1 = train.groupby(["housing","loan"]).agg(mean_balance=("balance","mean"),fd_rate=("y","mean")).reset_index()
table_1
```

```python
hm_table = table_1.pivot_table(index="housing",columns="loan",values=["mean_balance","fd_rate"])
fig, hm = plt.subplots(1, 2, figsize=(15, 5))

sns.heatmap(hm_table["mean_balance"], annot=True, fmt=".2f", cmap="Blues", vmin=0, vmax=2000, ax=hm[0])
hm[0].set_title("mean_balance")

sns.heatmap(hm_table["fd_rate"], annot=True, fmt=".3f", cmap="Reds", ax=hm[1])
hm[1].set_title("fd_rate")
hm[1].yaxis.set_visible(False)

plt.show()
```

## 2. 教育水準と年間平均残高の関係がどうやら職種ごとに違うように思う, 調べて欲しい

```python
table_2 = train[["id","education","balance","job"]]
hm_table_2 = table_2.groupby(["job","education"]).balance.mean().reset_index().pivot_table(columns="education",index="job",values="balance")
sns.heatmap(hm_table_2, annot=True, fmt=".2f",cmap="Reds")
```

```python
plt.figure(figsize=(15,5))
sns.catplot(table_2,x="job",y="balance",hue="education",kind="boxen",height=6,aspect=2,dodge=True,\
            hue_order=["primary", "secondary", "tertiary", "unknown"]).set_xticklabels(rotation=30)
```

## 3. コスト削減を目的として, 年齢でマーケティング対象を絞った場合に,
* Precision：マーケティング対象の契約率と
$$ \frac{TP}{(TP+FP)} $$
* Recall：契約者の捕捉率はどのようになるか？
$$ \frac{TP}{(TP+FN)} $$

```python
table_3 = train[["id","age","y"]].copy()
print("Y rate:", table_3["y"].mean())
# y rate by age
table_3_y = table_3.groupby("age").y.mean().reset_index()
sns.lineplot(data=table_3_y,x="age",y="y")
```

```python
age_filter = pd.DataFrame()
age_filter["age"] = train["age"].unique()
age_filter["precision"] = age_filter["age"].apply(lambda x: precision_score(train["y"],(train["age"] >= x).astype(int)))
age_filter["recall"] = age_filter["age"].apply(lambda x: recall_score(train["y"],(train["age"] >= x).astype(int)))
sns.lineplot(age_filter,x="age",y="precision",color="blue",label="Precision")
sns.lineplot(age_filter,x="age",y="recall",color="red",label="Recall")
```

```python
age_filter["precision"] = age_filter["age"].apply(lambda x: precision_score(train["y"],(train["age"] <= x).astype(int)))
age_filter["recall"] = age_filter["age"].apply(lambda x: recall_score(train["y"],(train["age"] <= x).astype(int)))
sns.lineplot(age_filter,x="age",y="precision",color="blue",label="Precision")
sns.lineplot(age_filter,x="age",y="recall",color="red",label="Recall")
```

```python
## Filtering age > 60 or < 30?
table_3["filter"] = ((table_3["age"] > 60) | (table_3["age"] < 30)).astype(int)
precision = precision_score(table_3["y"],table_3["filter"])
recall = recall_score(table_3["y"],table_3["filter"])
print(f"Precision: {precision:.3f}; Recall: {recall:.3f}")
```
