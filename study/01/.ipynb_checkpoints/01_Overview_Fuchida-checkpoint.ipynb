```python
# 課題3 コスト削減を目的として, 年齢でマーケティング対象を絞った場合に ,PrecisionとRecallはどのようになるか？
```

```python
# 年齢のしきい値（10〜100歳）
target_age_min = 10
target_age_max = 100

# マーケティング対象者の抽出
target = train[(train['age'] >= target_age_min) & (train['age'] <= target_age_max)]

# Precision: 対象者の中で契約した割合
precision = (target['y'] == 1).mean()

# Recall: 全契約者のうち、対象者に含まれていた割合
recall = train[(train['y'] == 1) & (train['age'] >= target_age_min) & (train['age'] <= target_age_max)].shape[0] / train[train['y'] == 1].shape[0]

print(f"Precision（契約率）: {precision:.3f}")
print(f"Recall（契約者の捕捉率）: {recall:.3f}")
```

```python
# 年齢のしきい値（50〜100歳）
target_age_min = 50
target_age_max = 100

# マーケティング対象者の抽出
target = train[(train['age'] >= target_age_min) & (train['age'] <= target_age_max)]

# Precision: 対象者の中で契約した割合
precision = (target['y'] == 1).mean()

# Recall: 全契約者のうち、対象者に含まれていた割合
recall = train[(train['y'] == 1) & (train['age'] >= target_age_min) & (train['age'] <= target_age_max)].shape[0] / train[train['y'] == 1].shape[0]

print(f"Precision（契約率）: {precision:.3f}")
print(f"Recall（契約者の捕捉率）: {recall:.3f}")
```

```python
# 年齢のしきい値（70〜100歳）
target_age_min = 70
target_age_max = 100

# マーケティング対象者の抽出
target = train[(train['age'] >= target_age_min) & (train['age'] <= target_age_max)]

# Precision: 対象者の中で契約した割合
precision = (target['y'] == 1).mean()

# Recall: 全契約者のうち、対象者に含まれていた割合
recall = train[(train['y'] == 1) & (train['age'] >= target_age_min) & (train['age'] <= target_age_max)].shape[0] / train[train['y'] == 1].shape[0]

print(f"Precision（契約率）: {precision:.3f}")
print(f"Recall（契約者の捕捉率）: {recall:.3f}")
```

```python
#　回答（一般論）　
# - Precision は上がる可能性が高い → 年齢層を絞ることで、契約率の高い層だけにアプローチできるため。
# - Recall は下がる可能性がある → 契約する可能性がある人を年齢で除外してしまうため。
```

```python
# 年齢グループ別申し込み率
```

```python
# 年齢をグループ化（10歳刻み）
train['age_group'] = pd.cut(train['age'], bins=[20,30,40,50,60,70,80], right=False)

# グループごとの申し込み率を計算
age_rate = train.groupby('age_group')['y'].apply(lambda x: (x == 1).mean())

# 可視化
age_rate.plot(kind='bar', color='steelblue')
plt.title('年齢グループ別の申し込み率')
plt.ylabel('申し込み率')
plt.xlabel('年齢グループ')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 前キャンペーン接触後の日数別
```

```python
# pdaysをビン分け（例：0, 1〜10, 11〜50, 51〜100, 101〜999）
bins = [-1, 0, 10, 50, 100, 999]
labels = ['未接触', '1-10日', '11-50日', '51-100日', '101日以上']
train['pdays_bin'] = pd.cut(train['pdays'], bins=bins, labels=labels)

# ビンごとの申し込み率を計算
pdays_rate = train.groupby('pdays_bin')['y'].apply(lambda x: (x == 1).mean())

# 可視化
pdays_rate.plot(kind='bar', color='teal')
plt.title('前キャンペーン接触後の日数別の申し込み率')
plt.ylabel('申し込み率')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
