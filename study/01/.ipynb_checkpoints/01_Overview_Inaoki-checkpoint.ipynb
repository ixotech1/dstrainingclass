# 250826_day2課題

```python
job_labels = {
    "admin.": "事務職",
    "blue-collar": "ブルーカラー",
    "entrepreneur": "起業家",
    "housemaid": "家政婦",
    "management": "管理職",
    "retired": "退職者",
    "self-employed": "自営業",
    "services": "サービス業",
    "student": "学生",
    "technician": "技術者",
    "unemployed": "失業者",
    "unknown": "不明"
}
edu_labels = {
    "primary": "初等教育",
    "secondary": "中等教育",
    "tertiary": "高等教育",
    "unknown": "不明"
}

train["job_ja"] = train["job"].map(job_labels)
train["education_ja"] = train["education"].map(edu_labels)
```

## 1. 住宅ローン/個人ローンごとの年間平均残高の平均/定額預金申込率

```python
train.groupby(["housing","loan"])[["balance","y"]].agg(
    年間平均残高の平均 = ("balance","mean"),
    定額預金申込率 = ("y","mean")
).reset_index()
```

```python
sns.catplot(data=train,x="housing",y="balance",kind="bar",hue="loan",hue_order=["yes","no"]).set(xlabel="住宅/個別ローンの有無", ylabel="年間平均残高")
sns.catplot(data=train,x="housing",y="y",kind="bar",hue="loan",hue_order=["yes","no"]).set(xlabel="住宅/個別ローンの有無",ylabel="定額預金申込率")
```

## 2. 職種ごとの「教育水準と年間平均残高の関係」

```python
train.groupby(["job_ja","education_ja"])["balance"].agg(
    年間平均残高の平均="mean"
)
```

```python
order = (train.loc[train["job_ja"] != "不明"].groupby("job_ja")["balance"].mean().sort_values(ascending=True).index.tolist())
order.append("不明")
sns.catplot(data=train,x="job_ja",y="balance",kind="bar",hue="education_ja",order=order,hue_order=["初等教育","中等教育","高等教育","不明"],aspect=2).set(xlabel="職業", ylabel="年間平均残高")
plt.xticks(rotation=30);
```

```python
train.groupby(["job_ja","education_ja"])["y"].agg(
    定額預金申込率="mean",
    件数="count"
)
```

```python
order = (train.loc[train["job_ja"] != "不明"].groupby("job_ja")["y"].mean().sort_values(ascending=False).index.tolist())
order.append("不明")
sns.catplot(data=train,x="job_ja",y="y",kind="bar",hue="education_ja",order=order,hue_order=["初等教育","中等教育","高等教育","不明"],aspect=2).set(xlabel="職業", ylabel="年間平均残高")
plt.xticks(rotation=30);
```

## 3. 年齢閾値でのPrecision/Recall

```python
from sklearn.metrics import precision_score, recall_score
y_true = train["y"].astype(int)
thresholds = range(15, 96, 5)

precisions = [precision_score(y_true,(train["age"] >= T).astype(int)) for T in thresholds]
recalls = [recall_score(y_true,(train["age"] >= T).astype(int)) for T in thresholds]
sns.lineplot(x=thresholds, y=precisions, label="Precision（T歳以上の契約者の割合）", marker="o")
sns.lineplot(x=thresholds, y=recalls, label="Recall（T歳以上の契約者の全契約者数に対する割合）", marker="o");
```
